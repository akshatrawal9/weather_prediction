```python
import numpy as np

# Define states
states = ['Sunny', 'Cloudy', 'Rainy']
transition_matrix = np.array([[0.8, 0.1, 0.1],  # From Sunny
                              [0.2, 0.6, 0.2],  # From Cloudy
                              [0.3, 0.4, 0.3]]) # From Rainy

# Function to predict the next state
def predict_next_state(current_state):
    current_index = states.index(current_state)
    next_state_probabilities = transition_matrix[current_index]
    return np.random.choice(states, p=next_state_probabilities)

# Function to simulate weather over a number of days
def simulate_weather(start_state, days):
    weather_sequence = [start_state]
    current_weather = start_state

    for _ in range(days):
        next_weather = predict_next_state(current_weather)
        weather_sequence.append(next_weather)
        current_weather = next_weather

    return weather_sequence

# Example usage
initial_weather = 'Cloudy'
days_to_simulate = 7
weather_forecast = simulate_weather(initial_weather, days_to_simulate)

# Print the weather forecast
print(f'Initial Weather: {initial_weather}')
print('Weather Forecast for the next 7 days:')
for day, weather in enumerate(weather_forecast):
    print(f'Day {day}: {weather}')
```
